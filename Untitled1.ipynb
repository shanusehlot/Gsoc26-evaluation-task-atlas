```python
# Patch model.py for minGRU import (try/except)
import re
with open("model.py", "r") as f:
    content = f.read()
pattern = r'from minGRU_pytorch\.minGRU import \(.*?\)'
replacement = """try:
    from minGRU_pytorch.minGRU import (
        heinsen_associative_scan_log,
        log_g as _mingru_log_g,
        g as _mingru_g
    )
    _HAS_MINIGRU = True
except ImportError:
    _HAS_MINIGRU = False
    heinsen_associative_scan_log = None
    _mingru_log_g = None
    _mingru_g = None"""
new_content = re.sub(pattern, replacement, content, flags=re.DOTALL)
with open("model.py", "w") as f:
    f.write(new_content)
print("вњ… minGRU patch applied")

# Patch model.py for GRU training mode
with open("model.py", "r") as f:
    content = f.read()
pattern = r'(class GRUBlock.*?\n.*?def forward\(self, x, inference_params=None\):\s*\n)(.*?)(?=\n\s+def |\nclass |$)'
def repl(match):
    header = match.group(1)
    body = match.group(2)
    new_body = "        if self.training:\n            self.gru.train()\n" + body
    return header + new_body
new_content = re.sub(pattern, repl, content, flags=re.DOTALL)
with open("model.py", "w") as f:
    f.write(new_content)
print("вњ… GRU training mode patch applied")

# Patch train.py to restore train mode after validation
with open("train.py", "r") as f:
    content = f.read()
pattern = r'(val_bpp = evaluate_bpp\(model, val_loader, criterion, device=device, vocab_size=vocab_size\)\s*\n)'
def repl2(match):
    return match.group(1) + "        model.train()  # restore train mode after validation\n"
new_content = re.sub(pattern, repl2, content)
with open("train.py", "w") as f:
    f.write(new_content)
print("вњ… train.py patch applied")
```
